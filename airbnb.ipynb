### Occupancy Trend (Heatmap)
group (x-axis) --> month 

variable (y-axis) --> neighborhood

value (count) --> number of Airbnbs booked (value of false for "availability")

```python
calendar_grouped
```

```python
calendar_grouped['month'] = pd.to_datetime(calendar_grouped['date']).dt.month
calendar_grouped
```

```python
merge_this = listings[['id', 'neighbourhood']]
merge_this
```

```python
calendar_grouped['listing_id'] = calendar_grouped['listing_id'].astype(int)
merge_this['id'] = merge_this['id'].astype(int)
```

```python
# calendar_grouped.(merge_this, left_on = 'listing_id', right_on = 'id')

# occupancy_df = pd.merge(calendar_grouped, merge_this, left_on='listing_id', right_on='id', how='outer')

# merged_df = pd.merge(listings, grouped, left_on='id', right_on='listing_id', how='inner')


occupancy_df = pd.merge(calendar_grouped, merge_this, left_on='listing_id', right_on='id', how='left')


occupancy_df
```

```python
occupancy_df.groupby('neighbourhood').count()
```

```python
# # Step 1: Convert 'available' column to boolean (False for booked, True for available)
# occupancy_df['available'] = occupancy_df['available'].map({'t': True, 'f': False})

# # Step 2: Calculate the number of bookings (False -> booked, True -> available)
# occupancy_df['booked'] = ~occupancy_df['available']

# # Step 3: Group by 'month' and 'neighborhood' and calculate the average number of bookings
# heatmap_df = occupancy_df.groupby(['month', 'neighbourhood'])['booked'].mean().reset_index()

# heatmap_df
```

```python
desired_neighborhoods = [
    'Mission Bay', 'Pacific Beach', 'La Jolla', 'Ocean Beach', 'North Hills',
    'East Village', 'Midtown', 'Loma Portal', 'Gaslamp Quarter', 'Balboa Park'
]
filtered_df = occupancy_df[occupancy_df['neighbourhood'].isin(desired_neighborhoods)]
heatmap_df = filtered_df.groupby(['month', 'neighbourhood'])['booked'].mean().reset_index()

# Step 5: Rename columns to match desired format
heatmap_df.columns = ['group', 'variable', 'value']
heatmap_df
```

```python
heatmap_df.to_csv('heatmap_df.csv')
```
